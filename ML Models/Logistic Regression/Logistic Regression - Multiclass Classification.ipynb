```python
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
from sklearn.datasets import load_digits
%matplotlib inline
```

```python
ld = load_digits()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(ld.data,ld.target, test_size=0.2)
```

```python
X_train # Here each data represents 1d array, so it forms 2d array
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
ld.target[[2, 3]]
```

```python
model.predict(ld.data[2:4])
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted) # parameters should be passed in order as (actual values, predicted_values)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
